```python
import pandas as pd
import numpy as np
```

```python
oil_data = pd.read_csv('output/oil_prices.csv')
oil_data
oil_data['Year and month'] = pd.to_datetime(oil_data['Date']).dt.to_period('M')
oil_data
```

```python
# Percent change by month
# need to make csv


copy2 = oil_data.copy()
copy2['Date'] = pd.to_datetime(copy2['Date']).dt.to_period('M')
copy2['Date']
copy2 = copy2.drop(columns=['Year and month'])
copy2.index
copy2 = copy2.set_index('Date')
copy2.index
copy2['Percent Change'] = copy2.pct_change(periods=1, freq='M')
copy2['Percent Change'] = copy2['Percent Change'].fillna(0)
copy2
copy2_group = copy2.groupby(pd.Grouper(freq='M'))
percent_ch = copy2_group['Percent Change'].sum()
av_price = copy2_group['Price'].mean()
av_price

clean_df = pd.DataFrame({
    'Average Price': av_price,
    'Percent Change(month)': percent_ch,
})
clean_df
```
