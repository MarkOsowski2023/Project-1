```python
import yfinance as yf
import pandas as pd
import numpy as np
import requests
import lxml
from pprint import pprint
```

```python
# Examples of Using yfinance

# For Tesla (ticker TSLA)
tsla = yf.Ticker('TSLA')

tsla.info
```

```python
# Dataframe of stock price history (exactly what we need)
tsla_prices_df = tsla.history(period='max')
tsla_prices_df['Year and Month'] = pd.to_datetime(tsla_prices_df.index)
tsla_prices_df['Filter Dates'] = tsla_prices_df['Year and Month'].dt.to_period('M')
tsla_prices_df
```

```python
tsla_group = tsla_prices_df.groupby('Filter Dates')
tsla_group['High'].mean()
```
