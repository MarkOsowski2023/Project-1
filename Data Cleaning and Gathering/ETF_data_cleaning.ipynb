```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
driv_dat = pd.read_csv('output/driv.csv')
driv_dat
driv_data = driv_dat.copy()
```

```python
# INDEXING
# making date time index's
# 1 is going to be yyyy-mm-dd
# 2 is going to be yyyy-mm
# 3 is going to be yyyy

# 1
driv_data['Date'] = pd.to_datetime(driv_data['Date'])
driv_data = driv_data.set_index('Date')
driv_data = driv_data.drop(columns=['Year Month Day'])
driv_data.reset_index(inplace=True)
driv_data
```

```python
# 2
copy_driv_dat = driv_dat.copy()
copy_driv_dat['Date'] = pd.to_datetime(copy_driv_dat['Date']).dt.to_period('M')
copy_driv_dat = copy_driv_dat.set_index('Date')
copy_driv_dat = copy_driv_dat.drop(columns=['Year Month Day'])
copy_driv_dat.reset_index(inplace=True)
copy_driv_dat
```

```python
# 3 
driv_3 = driv_dat.copy()
driv_3['Date'] = pd.to_datetime(driv_3['Date']).dt.to_period('Y')
driv_3 = driv_3.set_index('Date')
driv_3 = driv_3.drop(columns=['Year Month Day'])
driv_3.reset_index(inplace=True)
driv_3
```
