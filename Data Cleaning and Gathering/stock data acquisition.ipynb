```python
import yfinance as yf
import pandas as pd
import numpy as np
import requests
import lxml
from pprint import pprint
```

```python
# DRIV

# Global X Autonomous & Electric Vehicles ETF

# Broadly focused with a portfolio that includes tech giants, automakers, semiconductor companies, and other EV suppliers

driv = yf.Ticker('DRIV')
driv_df = driv.history(period='max')
driv_df
driv_df['Year Month Day'] = pd.to_datetime(driv_df.index)
driv_df['ETF name'] = 'DRIV'
driv_df = driv_df.reindex(columns=['ETF name', 'Year Month Day', 'Open', 'High', 'Low', 'Close', 'Volume'])
driv_df
```
