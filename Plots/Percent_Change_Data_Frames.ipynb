```python
import pandas as pd
import numpy as np
```

```python
# read csv
driv1 = pd.read_csv('output/driv.csv')
driv1.head()
```

```python
# Making Data Frame for Average ETF Price (High), Average Volume By Month and Year
# And Percent Change for ETF Price and Volume By Month and Year

# DRIV

# make copy
driv = driv1.copy()

# PERCENT CHANGE
# setting datetime index (yyyy-mm)
# percent change of Stock High for each index
driv['Date'] = pd.to_datetime(driv['Date']).dt.to_period('M')
driv = driv.drop(columns=['Year Month Day'])
driv = driv.set_index('Date')
driv
driv['High (Percent Change)'] = driv['High'].pct_change(periods=1, freq='M')
driv
driv['High (Percent Change)'] = driv['High (Percent Change)'].fillna(0)
driv['Volume (Percent Change)'] = driv['Volume'].pct_change(periods=1, freq='M')
driv['Volume (Percent Change)'] = driv['Volume (Percent Change)'].fillna(0)

# Group by (yyyy-mm)
driv_group = driv.groupby(pd.Grouper(freq='M'))
driv_pc = driv_group['High (Percent Change)'].sum()
driv_v_pc = driv_group['Volume (Percent Change)'].sum()
av_high = driv_group['High'].mean()
av_vol = driv_group['Volume'].mean()

# New Data Frame
driv_clean = pd.DataFrame({
    'ETF': "DRIV",
    'Average High': av_high,
    'High (Percent Change)': driv_pc,
    'Average Volume': av_vol,
    'Volume (Percent Change)': driv_v_pc,
})

# Reset Index
driv_clean.reset_index(inplace=True)
driv_clean.head()
```

```python
# to csv
driv_clean.to_csv('csv_files/driv_percent_change.csv', encoding='utf-8', index=False)
```
